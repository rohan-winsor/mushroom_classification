```python
import pandas as pd 
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
data_df = pd.read_csv("mushrooms.csv",index_col=False)
```

```python
data_df.head()
```

```python
label = data_df['class']
label.replace('p',0,inplace=True)
label.replace('e',1,inplace=True)
label.head()
```

```python
features = data_df.drop(columns=['class'])
features.head()
```

```python
features_number = pd.get_dummies(features)
print(features_number.shape)
features_number.head()
```

```python
features_values = features_number.values.astype('float32')
label_values = label.values.astype('float32')
```

```python
features_train, features_test,label_train,label_test = train_test_split\
(features_values, label_values, test_size=0.2)
features_train, features_validation,label_train,label_validation = train_test_split\
(features_train, label_train, test_size=0.2)
```

```python
l0 = keras.layers.Dense(32, input_shape=(117,))
l1 = keras.layers.Dense(20, activation=tf.nn.relu)
l2 = keras.layers.Dense(2, activation='softmax')
```

```python
model = keras.Sequential([l0,l1,l2])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['acc'])
```

```python
history = model.fit(features_train,label_train,epochs=10,validation_data=(features_validation,label_validation))
```

```python
predicted_feature = model.predict(features_test)
model.evaluate(features_test,label_test)
```
